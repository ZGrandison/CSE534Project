```python
from matplotlib import pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline
ping_values_tcp = []
jitter_values_tcp = []
x_axis_tcp = []
with open('TCPPlayer1.txt') as f:
    i = 0
    while True:
        line = f.readline()
        t = line.split(",")
        if t[0] != '':
            ping = int(t[0])
            jitter = float(t[1].strip('\n'))
            ping_values_tcp.append(ping)
            jitter_values_tcp.append(jitter)
            x_axis_tcp.append(i / 250)
            i += 1
        if not line:
            break
      
ping_values_udp = []
jitter_values_udp = []
x_axis_udp = []
with open('UDPPlayer1.txt') as f:
    i = 0
    while True:
        line = f.readline()
        t = line.split(",")
        if t[0] != '':
            ping = int(t[0])
            jitter = float(t[1].strip('\n'))
            ping_values_udp.append(ping)
            jitter_values_udp.append(jitter)
            x_axis_udp.append(i / 250)
            i += 1
        if not line:
            break
            
#print(ping_values)
plt.plot(x_axis_tcp, ping_values_tcp, label='TCP')
plt.plot(x_axis_udp, ping_values_udp, label='UDP')
plt.xlabel("Time (s)")
plt.ylabel("Ping (ms)")
plt.title("Base TCP vs UDP ping over 30s")
plt.legend(loc="upper right")
plt.show()


plt.plot(x_axis_tcp, jitter_values_tcp, label='TCP')
plt.plot(x_axis_udp, jitter_values_udp, label='UDP')
plt.xlabel("Time (s)")
plt.ylabel("Jitter (ms)")
plt.title("Base TCP vs UDP jitter over 30s")
plt.legend(loc="upper right")
plt.show()
```
