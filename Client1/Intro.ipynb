## An example: visualizing data in the notebook âœ¨

Below is an example of a code cell. We'll visualize some simple data using two popular packages in Python. We'll use [NumPy](https://numpy.org/) to create some random data, and [Matplotlib](https://matplotlib.org) to visualize it.

Note how the code and the results of running the code are bundled together.

```python
from matplotlib import pyplot as plt
import numpy as np

# Generate 100 random data points along 3 dimensions
x, y, scale = np.random.randn(3, 100)
fig, ax = plt.subplots()

# Map each onto a scatterplot we'll create with Matplotlib
ax.scatter(x=x, y=y, c=scale, s=np.abs(scale)*500)
ax.set(title="Some random data, created with JupyterLab!")
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline
ping_values_tcp = []
jitter_values_tcp = []
x_axis_tcp = []
with open('untitled.txt') as f:
    i = 0
    while True:
        line = f.readline()
        t = line.split(",")
        if t[0] != '':
            ping = int(t[0])
            jitter = float(t[1].strip('\n'))
            ping_values_tcp.append(ping)
            jitter_values_tcp.append(jitter)
            x_axis_tcp.append(i / 250)
            i += 1
        if not line:
            break
      
ping_values_udp = []
jitter_values_udp = []
x_axis_udp = []
with open('untitled1.txt') as f:
    i = 0
    while True:
        line = f.readline()
        t = line.split(",")
        if t[0] != '':
            ping = int(t[0])
            jitter = float(t[1].strip('\n'))
            ping_values_udp.append(ping)
            jitter_values_udp.append(jitter)
            x_axis_udp.append(i / 250)
            i += 1
        if not line:
            break
            
#print(ping_values)
plt.plot(x_axis_tcp, ping_values_tcp, label='TCP')
plt.plot(x_axis_udp, ping_values_udp, label='UDP')
plt.xlabel("Time (s)")
plt.ylabel("Ping (ms)")
plt.title("Base TCP vs UDP ping over 30s")
plt.legend(loc="upper right")
plt.show()


plt.plot(x_axis_tcp, jitter_values_tcp, label='TCP')
plt.plot(x_axis_udp, jitter_values_udp, label='UDP')
plt.xlabel("Time (s)")
plt.ylabel("Jitter (ms)")
plt.title("Base TCP vs UDP jitter over 30s")
plt.legend(loc="upper right")
plt.show()
```
